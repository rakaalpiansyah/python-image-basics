3. Eksplorasi Dasar dengan Python 
3.1 Membaca dan Menampilkan Gambar

```python
import cv2 
from matplotlib import pyplot as plt 

image = cv2.imread('image1.jpg') 

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 

plt.imshow(image_rgb)
plt.title('Gambar Awal') 
plt.axis('off') 
plt.show() 
```

Tugas 3.1: Gunakan gambar lain, coba tampilkan dalam warna grayscale

```python
image_other = cv2.imread('image2.jpg')
image_gray = cv2.cvtColor(image_other, cv2.COLOR_BGR2GRAY)

plt.imshow(image_gray, cmap='gray')
plt.title('Gambar Grayscale (Tugas 3.1)')
plt.axis('off')
plt.show()
```

3.2 Operasi Dasar pada Gambar

```python
resized_image = cv2.resize(image, (200, 200)) 

cropped_image = image[50:150, 50:150] 

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Resize') 

plt.subplot(1, 2, 2) 
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)) 
plt.title('Crop') 

plt.show() 
```

3.3 Membuat Histogram Gambar

```python
colors = ('b', 'g', 'r') 

for i, color in enumerate(colors):
    histogram = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(histogram, color=color)

plt.title('Histogram Warna')
plt.xlabel('Intensitas') 
plt.ylabel('Jumlah Piksel') 
plt.show() 
```

Tugas 3.4: Gunakan gambar grayscale dan bandingkan histogramnya dengan gambar berwarna.

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
for i, color in enumerate(('b', 'g', 'r')):
    hist_color = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(hist_color, color=color)
plt.title('Histogram Gambar Berwarna')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Piksel')

plt.subplot(1, 2, 2)
hist_gray = cv2.calcHist([image_gray], [0], None, [256], [0, 256])
plt.plot(hist_gray, color='black')
plt.title('Histogram Gambar Grayscale')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Piksel')

plt.tight_layout()
plt.show()
```

4. Diskusi dan Refleksi
Diskusikan hasil yang diperoleh dengan menjawab pertanyaan berikut: 
1. Apa perbedaan gambar asli, grayscale, dan hasil crop? 
    - Gambar Asli: Memiliki 3 saluran warna (RGB/BGR) yang merepresentasikan spektrum warna secara lengkap.
    - Grayscale: Hanya memiliki 1 saluran warna, di mana nilai pikselnya merepresentasikan tingkat intensitas cahaya dari hitam (0) hingga putih (255) tanpa spektrum warna.
    - Hasil Crop: Merupakan bagian spesifik dari gambar asli yang dipotong berdasarkan koordinat matriks tertentu. Ukuran resolusinya menjadi lebih kecil, namun sifat channel warnanya tetap sama dengan sumber aslinya.


2. Bagaimana histogram dapat memberikan informasi tentang distribusi warna pada gambar? 
    - Histogram menampilkan grafik frekuensi jumlah piksel (sumbu Y) terhadap tingkat intensitas cahayanya (sumbu X, dari 0-255). Dari histogram, kita bisa mengetahui apakah gambar terlalu gelap (puncak grafik menumpuk di area kiri/nilai rendah), terlalu terang (menumpuk di kanan), atau memiliki kontras yang baik (terdistribusi merata). Pada gambar berwarna, histogram juga menunjukkan intensitas dominan dari masing-masing warna Red, Green, dan Blue.

Tugas Mandiri

```python
gambar_mandiri = cv2.imread('image3.jpg')

# 1. Resize gambar menjadi 300x300
img_300 = cv2.resize(gambar_mandiri, (300, 300))

# 2. Crop bagian tengah gambar (mengambil area 100x100 di tengah)
img_tengah = img_300[100:200, 100:200]

# 3. Tampilkan histogram grayscale-nya
img_tengah_gray = cv2.cvtColor(img_tengah, cv2.COLOR_BGR2GRAY)
hist_mandiri = cv2.calcHist([img_tengah_gray], [0], None, [256], [0, 256])

# Visualisasi
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img_300, cv2.COLOR_BGR2RGB))
plt.title('Resize 300x300')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(img_tengah, cv2.COLOR_BGR2RGB))
plt.title('Crop Bagian Tengah')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.plot(hist_mandiri, color='black')
plt.title('Histogram Grayscale')
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Piksel')

plt.tight_layout()
plt.show()
```
